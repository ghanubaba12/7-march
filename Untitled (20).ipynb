```python
Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.
ans-
Measures of dispersion, such as range, variance, and standard deviation, are used to describe the spread of a dataset.

Range: It measures the difference between the maximum and minimum values in a dataset. A larger range indicates a greater spread, while a smaller range indicates a smaller spread. For example, if we have a dataset of test scores, the range would be the difference between the highest and lowest scores. A larger range would indicate that the test scores are more spread out, while a smaller range would indicate that the scores are more tightly clustered around the mean.

Variance: It measures how much the individual data points deviate from the mean of the dataset. A larger variance indicates a greater spread, while a smaller variance indicates a smaller spread. For example, if we have a dataset of salaries, the variance would measure how much the individual salaries deviate from the average salary. A larger variance would indicate that the salaries are more spread out, while a smaller variance would indicate that the salaries are more tightly clustered around the mean.

Standard deviation: It measures the amount of variation or dispersion of a dataset relative to its mean. A larger standard deviation indicates a greater spread, while a smaller standard deviation indicates a smaller spread. For example, if we have a dataset of heights, the standard deviation would measure how much the individual heights vary from the average height. A larger standard deviation would indicate that the heights are more spread out, while a smaller standard deviation would indicate that the heights are more tightly clustered around the mean.

In summary, measures of dispersion are used to describe the degree to which individual data points deviate from the central tendency of a dataset. These measures help to provide information about the spread of the data, which can be useful for understanding the distribution and making informed decisions based on the data.




```

```python
Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
dataset.
The formula for calculating the sample mean is:

mean = (sum of all values in the sample) / (number of values in the sample)

For example, suppose we have the following dataset:

5, 10, 12, 15, 20

To calculate the sample mean, we first add up all the values in the sample:

5 + 10 + 12 + 15 + 20 = 62

Next, we divide the sum by the number of values in the sample, which in this case is 5:

62 / 5 = 12.4

Therefore, the sample mean for this dataset is 12.4.




```

```python
Q13. How is covariance different from correlation?
ans-
Covariance and correlation are both measures of the relationship between two variables. However, there are some key differences between the two:

Scale: Covariance measures the direction and strength of the linear relationship between two variables, but it does not tell us about the magnitude or the scale of the relationship. Correlation, on the other hand, measures both the strength and the direction of the linear relationship, and it is scaled between -1 and 1.

Interpretation: Covariance can be difficult to interpret because its value depends on the scale of the variables being measured. For example, if we are looking at the covariance between income and age, the value of the covariance will depend on the units in which income and age are measured. Correlation, on the other hand, is easier to interpret because its value is standardized and does not depend on the scale of the variables.

Range: Covariance can take on any value, positive, negative, or zero, while correlation is always between -1 and 1. A covariance of zero means that there is no linear relationship between the two variables, while a correlation of zero means that there is no linear relationship, and it also indicates that there is no non-linear relationship either.

In summary, correlation is a standardized version of covariance, and it is a more interpretable and widely used measure of the relationship between two variables.




```
